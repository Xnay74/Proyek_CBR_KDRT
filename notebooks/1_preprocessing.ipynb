```python
import os
import re
from pdfminer.high_level import extract_text

# 1. Tentukan jalur folder input (PDF) dan output (TXT) secara relatif
raw_pdf_dir = "../data/raw/"
output_txt_dir = "../data/raw_txt/"  # Folder baru untuk menyimpan teks bersih

# Buat folder output jika belum ada
os.makedirs(output_txt_dir, exist_ok=True)

# 2. Fungsi untuk membersihkan teks (case folding & hapus whitespace berlebih)
def clean_text(text):
    text = re.sub(r'\s+', ' ', text) # Merapikan spasi & enter yang berantakan
    text = text.lower()              # Mengubah semua huruf menjadi huruf kecil
    return text.strip()

# 3. Looping untuk memproses ke-60 file PDF kamu
print("Memulai proses konversi dan pembersihan teks secara lokal...\n")

list_files = [f for f in os.listdir(raw_pdf_dir) if f.endswith('.pdf')]

for index, file_name in enumerate(list_files, 1):
    pdf_path = os.path.join(raw_pdf_dir, file_name)
    
    try:
        # Ekstrak teks dari PDF menggunakan library yang baru diinstall
        raw_text = extract_text(pdf_path)
        
        # Bersihkan teksnya
        cleaned_text = clean_text(raw_text)
        
        # Simpan hasil ke format .txt dengan nama baru yang teratur (case_001.txt)
        new_file_name = f"case_{index:03d}.txt"
        txt_path = os.path.join(output_txt_dir, new_file_name)
        
        with open(txt_path, "w", encoding="utf-8") as txt_file:
            txt_file.write(cleaned_text)
            
        print(f"[{index}/{len(list_files)}] Sukses konversi: {file_name} -> {new_file_name}")
        
    except Exception as e:
        print(f" ❌ Gagal memproses {file_name}. Error: {e}")

print("\n🎉 Semua file selesai dikonversi ke format teks (.txt) di folder data/raw_txt/!")
```
